```python
import pandas as pd

# Load data
df = pd.read_csv("data/reviews.csv")

# Show first 5 rows
df.head()
```

```python
# Check data info
df.info()
```

```python
# Remove duplicate reviews
df = df.drop_duplicates()

print("After removing duplicates:", len(df))
```

```python
# Remove empty reviews
df = df.dropna()

print("After removing null values:", len(df))
```

```python
# Review length column create
df["review_length"] = df["review"].apply(len)

df.head()
```

```python
df["review_length"].describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["review_length"])
plt.title("Review Length Distribution")
plt.xlabel("Length")
plt.ylabel("Count")
plt.show()
```

```python
from textblob import TextBlob

# Sentiment function
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df["sentiment_score"] = df["review"].apply(get_sentiment)

df.head()
```

```python
df["sentiment_score"].hist()
plt.title("Sentiment Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# visualization folder create
os.makedirs("visualizations", exist_ok=True)

# clean column names
df.columns = df.columns.str.strip().str.lower()

print("Available columns:", df.columns.tolist())

# sentiment label create
if "sentiment_score" in df.columns:
    df["sentiment_label"] = df["sentiment_score"].apply(
        lambda x: "Positive" if x > 0 else "Negative"
    )

# create 2x2 layout
fig, axes = plt.subplots(2, 2, figsize=(14,10))
fig.suptitle("Airline Review Analysis Dashboard", fontsize=16)

# -----------------------------
# 1️⃣ Review Length
# -----------------------------
if "review_length" in df.columns:
    sns.histplot(df["review_length"], bins=30, ax=axes[0,0])
    axes[0,0].set_title("Review Length")
else:
    axes[0,0].text(0.5,0.5,"review_length not found", ha="center")

# -----------------------------
# 2️⃣ Sentiment Score
# -----------------------------
if "sentiment_score" in df.columns:
    sns.histplot(df["sentiment_score"], bins=20, ax=axes[0,1])
    axes[0,1].set_title("Sentiment Score")
else:
    axes[0,1].text(0.5,0.5,"sentiment_score not found", ha="center")

# -----------------------------
# 3️⃣ Positive vs Negative
# -----------------------------
if "sentiment_label" in df.columns:
    sns.countplot(x=df["sentiment_label"], ax=axes[1,0])
    axes[1,0].set_title("Sentiment Count")
else:
    axes[1,0].text(0.5,0.5,"sentiment data not found", ha="center")

# -----------------------------
# 4️⃣ Empty plot (future use)
# -----------------------------
axes[1,1].axis("off")
axes[1,1].set_title("More Analysis (Future)")

plt.tight_layout()
plt.savefig("visualizations/dashboard.png")
plt.show()
```
